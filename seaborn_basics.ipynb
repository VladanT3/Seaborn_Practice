```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_scores = pd.read_excel('histograms.xlsx')
df_scores.head()
```

```python
sns.histplot(data=df_scores, x='Exam_Score', kde=True)

plt.xlabel('Score')
plt.ylabel('Count')
plt.title('Exam Scores')
plt.show
```
